```python
import pandas as pd
import numpy as np
import scipy
import pywt
import pywt.data
import pylab
import matplotlib.pyplot as plt
deformed_data= pd.read_csv('E:\Thesis\Deformed data.csv')
from PIL import Image
import scipy.misc
import imageio
```

```python
sig= np.array(deformed_data[10 :])
sig.shape
```

```python
x= sig[:, 1:2]
x=x.reshape(-1)
x.shape
```

```python
l_frequency_Ref, m_frequency_Ref, h_frequency_Ref= np.split(x, [330,660])
```

```python
h_frequency_Ref.shape
```

```python
scales=np.arange(1, 101)
w = pywt.ContinuousWavelet('cmor1-1')
#w.bandwidth_frequency=2
#w.center_frequency=2
```

```python
plt.figure(figsize=(14,20))
ax1 = plt.subplot(311)
cfs1,freqs1 = pywt.cwt(l_frequency_Ref,scales,wavelet=w )
#ax1.imshow(cfs1.real, interpolation='nearest', aspect='auto' )
plt.imshow(cfs1.real, extent=[1, 330, 20, 2107], cmap='jet', aspect='auto',
            vmax=abs(cfs1.real).max(), vmin=-abs(cfs1.real).max())
#plt.imsave('1.jpg', cfs1.real)
scipy.misc.imsave('58.jpg', cfs1.real)
ax2 = plt.subplot(312)
cfs2,freqs2 = pywt.cwt(m_frequency_Ref,scales,wavelet=w )
plt.imshow(cfs2.real, extent=[331, 660,2130, 78566], cmap='jet', aspect='auto',
            vmax=abs(cfs1.real).max(), vmin=-abs(cfs1.real).max())
#plt.imsave('2.jpg', cfs2.real)
scipy.misc.imsave('59.jpg', cfs2.real)
ax3 = plt.subplot(313)
cfs3,freqs3 = pywt.cwt(h_frequency_Ref,scales,wavelet=w )
#ax3.imshow(cfs3.real, interpolation='nearest', aspect='auto' )
plt.imshow(cfs3.real, extent=[661, 990, 79432, 2000000], cmap='jet', aspect='auto',
            vmax=abs(cfs3.real).max(), vmin=-abs(cfs3.real).max())
#plt.imsave('3.jpeg', cfs3.real)
scipy.misc.imsave('60.jpg', cfs3.real)
```
